## Теоритическая справка
Пусть задана задача Коши в следующем виде: $$\begin{cases}
y\,'=f\,(x,y)   &x\in[a,b]\\
y(x_0)=y_0 &x_0 = a
\end{cases}$$  

Как и в предыдущих задачах, задано разбиение отрезка: $$ a = x_0 < x_1 < \cdots < x_n = b$$

Тогда приближенное решение в узлах $x_i$, которое обозначим через $y_i$, определяется по формуле (метод Эйлера): 
$$y_i = y_{i-1} + (x_i-x_{i-1})f(x_{i-1},y_{i-1}),\;\;\;\;\;\;         i=1,2,3,\ldots,n.$$

```python
import pandas as pd
import numpy as np
```

```python
# y' = x^2 - 2y
# y(x0) = 1
# [0;1], h = 0.1
def reshenie(x):
    return 3/4 * np.exp(-2*x) + 1/2 *x*x - 1/2 * x + 1/4


n = 10
a = 0
b = 1
h = (b - a) / n
X = [a]

for i in range(1, n+1):
    x = a + h * i
    #x = round(x, 3)
    X.append(x)
    
Y = [1]
# y[i] = y[i-1] + h * (f(xi, yi))


def f(x, y):
    return x*x - 2 * y


for i in range(n):
    y = Y[i] + h * f(X[i], Y[i])
    #y = round(y, 4)
    Y.append(y)
    
ans = [reshenie(X[i]) for i in range(n+1)]
delta = [np.abs(ans[i] - Y[i]) for i in range(n+1)]
```

```python
df = pd.DataFrame()
df['x[i]'] = X
df['y[i]'] = ans
df['y[i] by Euler'] = Y
df['sigma'] = delta
df
```

```python
def srav(delta, eps):
    for i in range(len(delta)):
        if delta[i] >= eps:
            return True
    return False
```

```python
while srav(delta, 0.01):
    n *= 2
    h = (b - a)/n
    X = [a]
    for i in range(1, n+1):
        x = a + h * i
        #x = round(x, 3)
        X.append(x)
    Y = [1]
    for i in range(n):
        y = Y[i] + h * f(X[i], Y[i])
        #y = round(y, 4)
        Y.append(y)
    ans = [reshenie(X[i]) for i in range(n+1)]
    delta = [np.abs(ans[i] - Y[i]) for i in range(n+1)]
```

```python
df = pd.DataFrame()
df['x[i]'] = X
df['y[i]'] = ans
df['y[i] by Euler'] = Y
df['sigma'] = delta
df
```
