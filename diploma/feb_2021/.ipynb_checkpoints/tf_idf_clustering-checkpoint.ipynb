# Text preprocessing

## Lower case

```python
texts = list(map(lambda x: x.lower(), texts))
texts[0]
```

## Remove punctuation, numbers, whitespaces and special symbols

```python
punct_to_remove = string.punctuation + '0123456789′'

def remove_punctuation(text):
    return text.translate(str.maketrans('', '', punct_to_remove))


def remove_whitespaces(text):
    return re.sub(' +', ' ', text)

texts = list(map(remove_punctuation, texts))
texts = list(map(remove_whitespaces, texts))
texts[0]
```

## Remove stopwords

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    return ' '.join([word for word in text.split() if word not in stop_words])

texts = list(map(remove_stopwords, texts))
texts[0]
```

## Remove frequent words

```python
from collections import Counter
cnt = Counter()
for text in texts:
    for word in text.split():
        cnt[word] += 1
        
cnt.most_common(10)
```

```python
freqwords = set([w for (w, wc) in cnt.most_common(10)])

def remove_freqwords(text):
    return ' '.join([word for word in text.split() if word not in freqwords])

texts = list(map(remove_freqwords, texts))
texts[0]
```

## Remove rare words

```python
rarewords = set([w for (w, wc) in cnt.most_common()[:-10-1:-1]])

rarewords
```

```python
def remove_rarewords(text):
    return ' '.join([word for word in text.split() if word not in rarewords])

texts = list(map(remove_rarewords, texts))
texts[0]
```

optionally we can: do stemming, lemmatization, spelling correction, but not necessary

# Clustering

## CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
import umap
import hdbscan
```

```python
def c_tf_idf(documents, m, ngram_range=(1, 1)):
    count = CountVectorizer(ngram_range=ngram_range, stop_words="english").fit(documents)
    t = count.transform(documents).toarray()
    w = t.sum(axis=1)
    tf = np.divide(t.T, w)
    sum_t = t.sum(axis=0)
    idf = np.log(np.divide(m, sum_t)).reshape(-1, 1)
    tf_idf = np.multiply(tf, idf)

    return tf_idf, count


def extract_top_n_words_per_topic(tf_idf, count, docs_per_topic, n=20):
    words = count.get_feature_names()
    labels = list(docs_per_topic.Topic)
    tf_idf_transposed = tf_idf.T
    indices = tf_idf_transposed.argsort()[:, -n:]
    top_n_words = {label: [(words[j], tf_idf_transposed[i][j]) for j in indices[i]][::-1] for i, label in enumerate(labels)}
    return top_n_words


def extract_topic_sizes(df):
    topic_sizes = (df.groupby(['Topic'])
                     .Doc
                     .count()
                     .reset_index()
                     .rename({"Topic": "Topic", "Doc": "Size"}, axis='columns')
                     .sort_values("Size", ascending=False))
    return topic_sizes
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)
umap_embeddings = umap.UMAP(n_neighbors=15,
                                n_components=5,
                                metric='cosine',
                                random_state=123).fit_transform(X)
```

```python
cluster = hdbscan.HDBSCAN(min_cluster_size=15,
                              metric='euclidean',
                              cluster_selection_method='eom').fit(umap_embeddings)
```

```python
umap_data = umap.UMAP(n_neighbors=15, n_components=2, min_dist=0.0, metric='cosine').fit_transform(X)
```

```python
labels = cluster.labels_

sil_score = silhouette_score(umap_embeddings[labels != -1], labels[labels != -1], metric='euclidean')
calinski_score = calinski_harabasz_score(umap_embeddings[labels != -1], labels[labels != -1])
davies_score = davies_bouldin_score(umap_embeddings[labels != -1], labels[labels != -1])

print(f'Number of topics = {np.unique(labels[labels != -1]).size}')
print(f'Number of outliers = {(labels == -1).sum()}')
print(f'silhouette score = {sil_score}')
print(f'calinski-harabasz score = {calinski_score}')
print(f'davies-bouldin score = {davies_score}')

result = pd.DataFrame(umap_data, columns=['x', 'y'])
result['labels'] = cluster.labels_

# Visualize clusters
fig, ax = plt.subplots(figsize=(20, 10))
clustered = result.loc[result.labels != -1, :]
plt.scatter(clustered.x, clustered.y, c=clustered.labels, s=10, cmap='hsv_r')
plt.colorbar()
plt.show()

docs_df = pd.DataFrame(texts, columns=["Doc"])
docs_df['Topic'] = cluster.labels_
docs_df['Doc_ID'] = range(len(docs_df))
docs_per_topic = docs_df.groupby(['Topic'], as_index = False).agg({'Doc': ' '.join})
tf_idf, count = c_tf_idf(docs_per_topic.Doc.values, m=len(data), ngram_range=(1, 2))
top_n_words = extract_top_n_words_per_topic(tf_idf, count, docs_per_topic, n=20)
topic_sizes = extract_topic_sizes(docs_df); topic_sizes.head(10)
```
